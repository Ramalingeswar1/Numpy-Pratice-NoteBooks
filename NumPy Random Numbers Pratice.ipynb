# Random Data Distribution

```python
#Data Distribution is a list of all possible values,and how often each value occurs.
#The random module offers methods that returns randomly generated data distributions
```

```python
#A random distribution is a set of Random numbers that a certain random density function.
#Probability Density Function: A function that describes a continuous probability. i.e. probability of all values in an array.
```

```python
#We can generate random numbers based on defined probabilities using the choice() method of the random module.

#The choice() method allows us to specify the probability for each value.

#he probability is set by a number between 0 and 1, where 0 means that the value will never occur and 1 means that the value 
#will always occur.

#Probabilities to be summed to 1 
```

```python
x=rd.choice([1,2,3,4],p=[0.0,0.3,0.5,0.2],size=(100))
print(x)#The sum of all probabilities should be summed to 1.
```

# Random Permutation

```python
#Random Permutation of elements

#Permuatation refers to arrangement of elements that describes the possibility of arrangements(no of ways)

#Random module provides two methods for this:
#1)shuffle() 2)permutation()
```

```python
#shuffle means shuffling of arrangement of elements in the array itself
arr=np.array([1,2,3,4])
rd.shuffle(arr)
print(arr)#the changes will be made to original array the result will not be stored
```

```python
#2)permutaion() method will generate random permutations of the provided array
arr=np.array([1,2,3,4,5])
arr1=rd.permutation(arr)#Permutation method will not make any changes to original array it will just return the modified array.
print(arr1)
```

# Normal (Gaussian) Distribution

```python
#The Normal Distribution is also called as Gaussian Distribution named after the scientist Gauss
#It fits the probability distribution of many events,i.e;IQ Scores,HeartBeat etc..

#use random.normal() method to get a normal distribution
#random.normal() has three parameters--->
#1)loc-------(Mean)where peak of bell exists
#2)scale-----(standard deviation)how flat the graph distribution should be
#3)size------The shape of the array
```

```python
x=rd.normal(loc=1,scale=2,size=(2,3))
print(x)
```

# Binomial Distribution

```python
#Binomial Distribution is a Discrete Distribution
#It describes the outcome of binary scenarios e.g:toss of a coin i.e;either head or tail
```

```python
#Binary Scenarios
"""
 some binary scenarios:

Coin Toss:
Scenario: Tossing a fair coin.
Binary Outcome: Heads (True) or Tails (False).
Weather Forecast:
Scenario: Predicting whether it will rain tomorrow.
Binary Outcome: Rain (True) or No Rain (False).
Medical Test Result:
Scenario: Conducting a medical test for a disease.
Binary Outcome: Positive (True) if the disease is detected, Negative (False) if not.
Light Switch:
Scenario: Turning on a light switch.
Binary Outcome: Light On (True) or Light Off (False).
Pass/Fail Test:
Scenario: Taking a pass/fail exam.
Binary Outcome: Pass (True) or Fail (False).
Presence Detection:
Scenario: Detecting if someone is present in a room.
Binary Outcome: Present (True) or Absent (False).
Email Spam Detection:
Scenario: Classifying emails as spam or not spam.
Binary Outcome: Spam (True) or Not Spam (False).
Car Alarm Activation:
Scenario: Detecting if a car alarm has been activated.
Binary Outcome: Alarm Activated (True) or Not Activated (False).
"""
```

```python
#It has three parameters:

#n - number of trials.

#p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).

#size - The shape of the returned array.

#Discrete Distribution:The distribution is defined at separate set of events, e.g. a coin toss's result is discrete as it can be
#only head or tails whereas height of people is continuous as it can be 170, 170.1, 170.11 and so on.
```

```python
x=rd.binomial(n=2,p=0.5,size=10)
print(x)
```

```python
#Visualisation of Binomial Distribution
sns.kdeplot(rd.binomial(n=2,p=0.5,size=10))
plt.show()
```

```python
#Difference Between Normal and Binomial Distribution
#The main difference is that normal distribution is continous whereas binomial is discrete, but if there are enough data points
#it will be quite similar to normal distribution with certain loc and scale.
```

```python
sns.kdeplot(rd.normal(loc=50,scale=5,size=1000),label='normal')
sns.kdeplot(rd.binomial(n=100,p=0.5,size=1000),label='binomial')
plt.show()
```

# Poisson Distribution

```python
#Poision Distribution is a Discrete Distribution.
#Possion Distribution estimates how many times an event can happen.e.g..If a person eats twice a day what is the probability 
#that he will eat thrice a day

#It has two parameters
#1)lam -rate or known number of occurences e.g.2 for above problem---------->lam means lambda
#2)size-shape of the returned array
```

```python
x=rd.poisson(lam=2,size=10)
print(x)#The array individual item represents occurence of each item
```

```python
#Visualisation of poisson distribution
sns.kdeplot(rd.poisson(lam=2,size=10))
plt.show()
```

# Uniform Distribution

```python
#uniform distribution used to describe probability where event has equal chances of occuring.
#e.g.generation of random numbers

#It has three parameters:

#a - lower bound - default 0 .0.

#b - upper bound - default 1.0.

#size - The shape of the returned array.
```

```python
from numpy import random

x = random.uniform(size=(2, 3))

print(x)
```

```python
#Visualisation of Uniform Distribution
sns.kdeplot(random.uniform(size=(3,3)))
plt.show()
```

# Logistic Distribution

```python
#Logistic Distribution is used to describe growth
#Logistic Distribution is extensively used in machine learning logistic regression,neural networks etc,,
#It has three parameters:

#loc - mean, where the peak is. Default 0.

#scale - standard deviation, the flatness of distribution. Default 1.

#size - The shape of the returned array.
```

```python

x = random.logistic(loc=1, scale=2, size=(2, 3))

print(x)
```

```python
sns.kdeplot(rd.logistic(size=1000))

plt.show()
```

# Multinomial Distribution

```python
#Multinomial distribution is a generalization of binomial distribution.

#It describes outcomes of multi-nomial scenarios unlike binomial where scenarios must be only one of two. e.g. Blood type of a population, dice roll outcome.

#It has three parameters:

#n - number of possible outcomes (e.g. 6 for dice roll).

#pvals - list of probabilties of outcomes (e.g. [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] for dice roll).

#size - The shape of the returned array.


x = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

print(x)
```

# Exponential Distribution

```python
#Exponential distribution is used for describing time till next event e.g. failure/success etc.

#It has two parameters:

##scale - inverse of rate ( see lam in poisson distribution ) defaults to 1.0.

#size - The shape of the returned array.
```

```python
x = random.exponential(scale=2, size=(2, 3))

print(x)
```

# Chi Square Distribution

```python
##Chi Square distribution is used as a basis to verify the hypothesis.

#It has two parameters:

#df - (degree of freedom).

#size - The shape of the returned array.

#Draw out a sample for chi squared distribution with degree of freedom 2 with size 2x3:

x = random.chisquare(df=2, size=(2, 3))
print(x)
```

```python
sns.kdeplot(random.chisquare(df=1, size=1000))

plt.show()
```
